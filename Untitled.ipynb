Match Summary

```python
import pandas as pd
import json
```

```python
with open('t20_wc_match_results.json') as f:
    data=json.load(f)
len(data)
```

```python
print(data[0]['matchSummary'])
```

```python
df_data=pd.DataFrame(data[0]['matchSummary'])
df_data.head()
```

```python
df_data.shape
```

```python
df_data.rename({'scorecard':'match_Id'},axis=1,inplace=True)
```

```python
df_data.head()
```

```python
match_ids_dict={}
for index,row in df_data.iterrows():
    key1=row['team1']+' Vs '+row['team2']
    key2=row['team2']+' Vs '+row['team1']
    match_ids_dict[key1]=row['scorecard']
    match_ids_dict[key2]=row['scorecard']
match_ids_dict
```

Batting Summary

```python
with open('t20_wc_batting_summary.json') as f:
    data=json.load(f)
    all_records=[]
    for rec in data:
        all_records.extend(rec['battingSummary'])
all_records
```

```python
df_batting=pd.DataFrame(all_records)
```

```python
df_batting['out/not_out']=df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not_out")
```

```python
df_batting.drop(columns=["dismissal"],inplace=True)
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x: x.replace("â€"," "))
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x:x.replace("\xa0"," "))
```

```python
df_batting['match_id']=df_batting['match'].map(match_ids_dict)
df_batting.head(12)
```

```python
df_batting.to_csv('batting_summary.csv',index=False)
```
